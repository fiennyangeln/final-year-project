```python
import numpy as np
import math as mt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time 
import os    
import copy
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
from numpy import linalg as LA
import matplotlib.pyplot as plt
from mainAutocnnBatchNorm import autocnn2
from sklearn.model_selection import train_test_split
```

```python
# prepare data
print('==> Preparing the data')
classes     = 10
no_of_epoch = 1000
train_data=torch.load('data/train_data.pt')
train_label=torch.load('data//train_label.pt')
test_data=torch.load('data//test_data.pt')
test_label=torch.load('data//test_label.pt')
x_train = torch.tensor(train_data,dtype=torch.float64)
y_train = torch.tensor(train_label,dtype=torch.int64)
nData,_,_,_ = x_train.shape
shuffled_indices = torch.randperm(nData)
nTraining = 0.9*nData
x_train = x_train[shuffled_indices]
y_train = y_train[shuffled_indices]
x_val = x_train[int(0.9*nData):nData+1]
y_val = y_train[int(0.9*nData):nData+1]
#x_train = x_train[0:int(0.9*nData)]
#y_train = y_train[0:int(0.9*nData)]
x_test = torch.tensor(test_data,dtype=torch.float64)
y_test = torch.tensor(test_label,dtype=torch.int64)
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
print(x_test.shape, y_test.shape)
```

```python
(ArAsNet, fccLayerEvo, convLayerEvo, Bias2, Var, convFeatureCorr, Error_test, Error_val, Error_train,
           Loss_test, Loss_val, Loss_train, Iter) = autocnn2(x_train, y_train, x_val, y_val, x_test, y_test, 
                                                         no_of_epoch, gpu = 'cuda:1')
```

```python
SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=8)                   # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(22, 14))

ax1 = axes[0,0]
ax2 = axes[0,1]
ax3 = axes[0,2]
ax4 = axes[1,0]
ax5 = axes[1,1]
ax6 = axes[1,2]

ax1.plot(Iter[0:len(convFeatureCorr)],convFeatureCorr,'k-')
ax1.set_title('Feature Convolution Score')
ax1.set_ylabel('$FCS$')
ax1.set_xlabel('Number of epochs')
ax1.yaxis.tick_right()
ax1.autoscale_view('tight')

ax2.plot(Iter[0:len(convLayerEvo)],convLayerEvo[0:len(convLayerEvo)],'k-')
ax2.set_title('Convolutional Stage Evolution')
ax2.set_ylabel('Number of Convolutional Stages')
ax2.set_xlabel('Number of epochs')
ax2.yaxis.tick_right()
ax2.autoscale_view('tight')

ax3.plot(Iter[0:len(Error_train)],Error_train,'g--', label='Training Error')
ax3.plot(Iter[0:len(Error_val)],Error_val,'r-.', label='Validation Error')
ax3.plot(Iter[0:len(Error_test)],Error_test,'k-', label='Testing Error')
ax3.legend(loc=(0.5, 0.6))
ax3.yaxis.tick_right()
ax3.set_title('Error of Training, Validation and Testing Phases')
ax3.set_ylabel('$\%$ of Error')
ax3.set_xlabel('Number of epochs')
ax3.autoscale_view('tight')

ax4.plot(Iter[0:len(Bias2)],Bias2,'k-', label='Bias')
ax4.plot(Iter[0:len(Var)],Var,'r--', label='Variance')
ax4.legend(loc=(0.6, 0.1))
ax4.set_ylabel('$Bias-Var$ Level')
ax4.set_title('Bias and Variance')
ax4.set_xlabel('Number of epochs')
ax4.yaxis.tick_right()
ax4.autoscale_view('tight')

ax5.plot(Iter[0:len(fccLayerEvo)],fccLayerEvo[0:len(fccLayerEvo)], 'k-')
ax5.set_title('Fully Connected Layer Evolution')
ax5.set_ylabel('Number of Fully Connected Layers')
ax5.set_xlabel('Number of epochs')
ax5.yaxis.tick_right()
ax5.autoscale_view('tight')

#ax6.plot(Iter[0:len(layer1)],layer1,'y-', label='$1^{st}$ layer')
#ax6.plot(Iter[0:len(layer2)],layer2,'r-.', label='$2^{nd}$ layer')
#ax6.plot(Iter[0:len(layer2)],layer3,'c-.', label='$3^{rd}$ layer')
ax6.set_title('Hidden Node Evolution')
ax6.set_ylabel('Number of Fully Hidden Nodes')
ax6.set_xlabel('Number of epochs')
ax6.legend(loc=(0.65, 0.1))
ax6.yaxis.tick_right()
ax6.autoscale_view('tight')
```
